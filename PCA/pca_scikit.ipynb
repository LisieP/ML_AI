# Using Scikit Learn built in PCA functionality
Import iris dataset with 150 flowers all belonging to one of 3 classes. Once the data is loaded, these species (classes) are simply 0, 1, and 2. The idea is to reproduce the image in the scikit-image website. 

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

iris_data = load_iris()
data = iris_data.data
labels = iris_data.target
classes = set(labels)
```

Each principal component captures some proportion of the total variance. The first principal component captures the most, the second captures the next most while being orthogonal to the first, and so on. The explained variance by a particular principal component is the amount of total variance that is contained along that component's axis. So we fit the data and investigate the explained variance on a graph

```python
pcaIris = PCA(n_components=4)
pcaIris.fit(data)
plt.plot(pcaIris.explained_variance_ratio_)
plt.show()
```

# Projecting the data.
Project the data onto the first two principal components. The idea is to see how much class separation is obtained by projecting onto the first two principal components.


```python
X_r = pcaIris.transform(data)

plt.scatter(X_r[:,0], X_r[:,1], c = labels)
plt.axis([-4, 4, -1.5,1.5])
plt.show()
```
